```python
def bgd_substract(image, slice_to_c = (50,300)):
        """ ATENTION: Substracts background, which is calculated using vertical strip at right side"""
        data_for_bgd_det = image[:,slice_to_c[0]:slice_to_c[1]]
        av = sum(data_for_bgd_det,1)/data_for_bgd_det.shape[1]
        return image - tile(av[:,newaxis],image.shape[1])
```
